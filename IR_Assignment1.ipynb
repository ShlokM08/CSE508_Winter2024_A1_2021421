```python
#Q1.1
!pip install contractions

from docx import Document
import os
import nltk
import string
import random

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import os
import nltk
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

from docx import Document
import os
import nltk
import string
import random

# Additional import for handling contractions
import contractions

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Downloading required NLTK data
nltk.download('punkt')
nltk.download('stopwords')

# Set of stopwords
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    # Expand contractions in the text
    text = contractions.fix(text)

    # Lowercase the text
    text = text.lower()

    # Tokenize the text
    tokens = word_tokenize(text)

    # Remove punctuation from each token
    table = str.maketrans('', '', string.punctuation)
    stripped = [w.translate(table) for w in tokens if w not in string.punctuation]

    # Remove remaining tokens that are not alphabetic, but keep alphanumeric to preserve numbers
    words = [word for word in stripped if (word.isalpha() or word.isdigit() or word.isalnum()) and word not in stop_words]

    return words

def process_files(input_directory, output_directory):
    print(f"Processing files in {input_directory}...")
    files = [f for f in os.listdir(input_directory) if f.endswith(".txt")]
    print(f"Total .txt files found: {len(files)}")

    for filename in files:
        input_file_path = os.path.join(input_directory, filename)
        output_file_path = os.path.join(output_directory, f"processed_{filename}")

        # Read and preprocess the text
        with open(input_file_path, 'r', encoding='utf-8') as file:
            text = file.read()
            words = preprocess_text(text)
            output_text = ' '.join(words)

        # Write the processed text, overwrite if file already exists
        with open(output_file_path, 'w', encoding='utf-8') as output_file:
            output_file.write(output_text)

        print(f"Processed file saved to {output_file_path}")

# Define input and output folder paths
input_folder_path = '/content/drive/My Drive/text_files'
output_folder_path = '/content/drive/MyDrive/post_processing_files'

process_files(input_folder_path, output_folder_path)


# Downloading required NLTK data
nltk.download('punkt')
nltk.download('stopwords')

# Set of stopwords
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    # Lowercase the text
    text = text.lower()

    # Tokenize the text
    tokens = word_tokenize(text)

    # Remove punctuation from each token
    table = str.maketrans('', '', string.punctuation)
    stripped = [w.translate(table) for w in tokens]

    # Remove remaining tokens that are not alphabetic and also remove stopwords
    words = [word for word in stripped if word.isalpha() and word not in stop_words]

    # Remove blank space tokens
    words = [word for word in words if word.strip() != '']

    return words

def process_files(input_directory, output_directory):
    print(f"Processing files in {input_directory}...")
    files = [f for f in os.listdir(input_directory) if f.endswith(".txt")]
    print(f"Total .txt files found: {len(files)}")

    for filename in files:
        input_file_path = os.path.join(input_directory, filename)
        output_file_path = os.path.join(output_directory, f"processed_{filename}")

        # Read and preprocess the text
        with open(input_file_path, 'r', encoding='utf-8') as file:
            text = file.read()
            words = preprocess_text(text)
            output_text = ' '.join(words)

        # Write the processed text, overwrite if file already exists
        with open(output_file_path, 'w', encoding='utf-8') as output_file:
            output_file.write(output_text)

        print(f"Processed file saved to {output_file_path}")

# Define input and output folder paths
input_folder_path = '/content/drive/My Drive/text_files'
output_folder_path = '/content/drive/MyDrive/post_processing_files'

process_files(input_folder_path, output_folder_path)
```

```python
#Q1.2
import os
import random

def print_file_content(file_path, label):
    with open(file_path, 'r', encoding='utf-8') as file:
        content = file.read()
    print(f"--- {label} ---\n{content[:500]}...")  # Print first 500 characters for brevity

def compare_files(input_directory, output_directory):
    # Get all text file names from the input directory
    files = [f for f in os.listdir(input_directory) if f.endswith(".txt")]
    # Select 5 random files
    selected_files = random.sample(files, min(len(files), 5))

    for filename in selected_files:
        original_file_path = os.path.join(input_directory, filename)
        processed_file_name = f"processed_{filename}"
        processed_file_path = os.path.join(output_directory, processed_file_name)

        # Check if the processed file exists
        if os.path.exists(processed_file_path):
            print(f"File: {filename}\n")
            print_file_content(original_file_path, "Before Processing")
            print_file_content(processed_file_path, "After Processing")
            print("\n" + "="*60 + "\n")
        else:
            print(f"Processed file for {filename} not found.")

# Define input and output folder paths
input_folder_path = '/content/drive/My Drive/text_files'
output_folder_path = '/content/drive/MyDrive/post_processing_files'

compare_files(input_folder_path, output_folder_path)
```

```python
#Q3f
import os
import pickle

class PositionalIndex:
    def __init__(self, directory):
        self.directory = directory
        self.index = self.build_index()

    def build_index(self):
        index = {}
        for file in os.listdir(self.directory):
            if file.endswith(".txt"):
                path = os.path.join(self.directory, file)
                with open(path, 'r', encoding='utf-8') as content:
                    for position, term in enumerate(content.read().split()):
                        postings = index.setdefault(term, {})
                        positions = postings.setdefault(file, [])
                        positions.append(position)
        return index

    def save(self, filepath):
        with open(filepath, 'wb') as file:
            pickle.dump(self.index, file)

    @staticmethod
    def load(filepath):
        with open(filepath, 'rb') as file:
            return pickle.load(file)

class PhraseQueryProcessor:
    def __init__(self, positional_index):
        self.positional_index = positional_index

    def find_documents(self, phrase):
        terms = preprocess_text(phrase)
        if not terms:
            return set()

        postings_lists = [self.positional_index.get(term, {}) for term in terms]
        if not all(postings_lists):
            return set()

        common_docs = set(postings_lists[0].keys())
        for postings in postings_lists:
            common_docs &= postings.keys()

        return self.validate_positions(common_docs, postings_lists, len(terms))

    def validate_positions(self, documents, postings_lists, term_count):
        valid_documents = set()
        for doc in documents:
            term_positions = [postings_lists[i][doc] for i in range(term_count)]
            for pos in term_positions[0]:
                if all((pos + i) in term_positions[i] for i in range(1, term_count)):
                    valid_documents.add(doc)
                    break
        return valid_documents


processed_files_folder = '/content/drive/My Drive/post_processing_files'
pos_index = PositionalIndex(processed_files_folder)

index_file = 'positional_index.pkl'
pos_index.save(index_file)
loaded_index = PositionalIndex.load(index_file)

query_processor = PhraseQueryProcessor(loaded_index)
query_count = int(input("Enter the number of queries: "))

for i in range(query_count):
    query_input = input(f"Enter phrase query {i+1}: ")
    result_docs = query_processor.find_documents(query_input)

    print(f"Documents retrieved for query {i+1}: {len(result_docs)}")
    print(f"Documents: {' '.join(result_docs)}")
```

```python
#q2
import os
import pickle
import nltk
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from docx import Document

nltk.download('punkt')
nltk.download('stopwords')
class InvertedIndex:
    def __init__(self, directory):
        self.directory = directory
        self.index = self.build_index()

    def build_index(self):
        index = {}
        for file in os.listdir(self.directory):
            if file.endswith(".txt"):
                path = os.path.join(self.directory, file)
                with open(path, 'r', encoding='utf-8') as content:
                    for term in content.read().split():
                        if term not in index:
                            index[term] = [file]
                        elif file not in index[term]:
                            index[term].append(file)
        return index

    def save(self, filepath):
        with open(filepath, 'wb') as file:
            pickle.dump(self.index, file)

    @staticmethod
    def load(filepath):
        with open(filepath, 'rb') as file:
            return pickle.load(file)



class QueryProcessor:
    def __init__(self, inverted_index):
        self.inverted_index = inverted_index

    def process(self, terms, operators):
        if not terms or not operators or not self.inverted_index:
            return set()

        result = set(self.inverted_index.get(terms[0], []))
        for i, operator in enumerate(operators):
            if i + 1 < len(terms):
                next_terms = terms[i + 1]
                postings = self.inverted_index.get(next_terms, [])
                if operator == 'AND':
                    result &= set(postings)
                elif operator == 'OR':
                    result |= set(postings)
                elif operator == 'AND NOT':
                    result -= set(postings)
                elif operator == 'OR NOT':
                    result = result.union(set(self.inverted_index.keys()) - postings)
        return result

    def execute_queries(self, num_queries):
      for i in range(num_queries):
          query_input = input(f"Enter query {i+1}: ")
          operator_input = input(f"Enter operations for query {i+1} separated by comma: ")
          terms = preprocess_text(query_input)
          operators = operator_input.split(',')

          if len(terms) != len(operators) + 1:
              print(f"Error: Number of operations must be one less than number of terms in the query.")
              continue

          query_display = ' '.join([f"{term} {op}" for term, op in zip(terms[:-1], operators)])
          query_display += f" {terms[-1]}"  # Add the last term separately

          results = self.process(terms, operators)

          print(f"Query {i+1}: {query_display}")
          print(f"Documents retrieved for query {i+1}: {len(results)}")
          print(f"Documents: {' '.join(results)}")



output_folder_path = '/content/drive/MyDrive/post_processing_files'

# Create an instance of InvertedIndex with the directory where your processed files are stored
index = InvertedIndex(output_folder_path)

# Now you can save the index
index_file = 'inverted_index.pkl'
index.save(index_file)

# Loading the saved index
loaded_index = InvertedIndex.load(index_file)  # This returns a dictionary

# Adjusted the line to pass the loaded_index dictionary directly
query_processor = QueryProcessor(loaded_index)

# Proceed with querying
num_queries = int(input("Enter the number of queries: "))
query_processor.execute_queries(num_queries)

```
